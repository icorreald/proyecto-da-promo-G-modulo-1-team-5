```python
import random 

class Juego_Piedra_Papel_Tijera:
    
    def __init__ (self): 
        
        self.reglas = {"piedra" : "papel" , "papel" : "tijera" , "tijera" : "piedra"} # --En el diccionario de reglas cada clave tiene asociado el valor que la vence en el juego de piedra, papel o tijeras
        self.lista = ['piedra', 'papel', 'tijera'] # --Definimos una lista con las opciones que aceptaremos en la entrada de datos
        self.puntos_jugadora = 0 # --Establecemos los puntos iniciales tanto para la jugadora como para la máquina
        self.puntos_maquina = 0
        self.ronda = 0
        
    def bienvenida (self): # --Método para hacer un print de bienvenida y explicación del juego
        
        self.banner1 = '--------------------------------------'
        self.banner2 = '------ PIEDRA, PAPEL O TIJERA -------'
        self.banner3 = 'Intenta ganar a la máquina'
        self.banner4 = "Reglas:"
        self.banner5 = "Papel vence a Piedra"
        self.banner6 = "Piedra vence a Tijera"
        self.banner7 = "Tijeras vence a Papel"
        print(self.banner1.center(40))
        print(self.banner2.center(40))
        print(self.banner3.center(40))
        print(f" ")
        print(self.banner4.center(40))
        print(self.banner5.center(40))
        print(self.banner6.center(40))
        print(self.banner7.center(40))
        print(f" ")
        
    def entrada_datos (self): # --Método con los inputs necesarios para jugar
        
        self.jugadora = input('Elige opción: Piedra, Papel o Tijera').lower().strip() # --Respuesta de la jugadora (lower para evitar mayúsculas y strip para evitar espacios no deseados antes/después del texto)
        self.maquina = random.choice(self.lista) # --Respuesta de la máquina (random.choice que va a buscar a la lista de elementos posibles definida en el init)
        
    def jugada (self): # --Método con la lógica tras una jugada
        
        if self.jugadora == self.maquina: # --Revisamos si ambas respuestas son iguales para ver si ha habido un empate. Si es así, no se asignarán puntos
            print(f"-------------- Ronda {self.ronda} --------------".center(40))
            print(f"¡Ha habido un empate!".center(40))
            print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
            print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
            print(f" ")
            
        else:
            
            if self.reglas.get(self.jugadora) == self.maquina: # --Usamos el input de la jugadora como clave para que nos devuelva su valor (recordamos: la opción que la vencería) y si este es igual a la respuesta de la máquina, la máquina gana la ronda
                self.puntos_maquina += 1 # --Como la máquina ha ganado, le sumamos 1 punto
                print(f"-------------- Ronda {self.ronda} --------------".center(40))
                print(f"¡Lástima! Has perdido...".center(40))
                print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
                print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
                print(f" ")
                
            else: # --Si la condición anterior no se cumple, significa que la jugadora ha ganado, ya que la máquina no ha obtenido la opción que podría vencerla
                self.puntos_jugadora += 1 # --Como la jugadora ha ganado, le sumamos 1 punto
                print(f"-------------- Ronda {self.ronda} --------------".center(40))
                print(f"¡Genial! ¡Has ganado!".center(40))
                print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
                print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
                print(f" ")
                
    def resultado (self): # --Método con la definición del resultado del juego
        
        if self.puntos_jugadora > self.puntos_maquina:
            print(f"¡¡HAS GANADO AL PIEDRA, PAPEL O TIJERA!!".center(40))
            print(f"Venciste a la máquina con un total de {self.puntos_jugadora} puntos".center(40))
            print(self.banner1.center(40))
            
        else:
            print(f"PERDISTE AL PIEDRA, PAPEL O TIJERA...".center(40))
            print(f"Te quedaste con un total de {self.puntos_jugadora} punto(s)".center(40))
            print(self.banner1.center(40))
            
        
    def jugar (self):
        
        self.bienvenida()
        
        while self.puntos_jugadora < 3 and self.puntos_maquina < 3:
            
            self.ronda += 1
        
            self.entrada_datos()
            
            if self.jugadora not in self.reglas:
                print('¡Vaya! Ha ocurrido un error. Prueba otra vez con Piedra, Papel o Tijera'.center(40))
                print(f" ")
                
            else: 
                self.jugada()
                
        self.resultado()     
```

```python
piedra_papel_tijera = Juego_Piedra_Papel_Tijera()
piedra_papel_tijera.jugar()
```
