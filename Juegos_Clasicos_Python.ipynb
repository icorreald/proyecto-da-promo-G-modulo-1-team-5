```python
class Juegos:
    
#! ---------------------------------------------------------- Prints: PREGUNTAS Y RESPUESTAS
    
    def trivial_banner_inicio(self):
        print(('-'*20).center(47))
        print('PREGUNTAS Y RESPUESTAS'.center(47))
        print(('-'*20).center(47))
        print(' ')
        print(('-'*10).center(47))
        print('Victoria: acierta 5 preguntas'.center(47))
        print('Derrota: falla 3 preguntas'.center(47))
        print('Tienes dos intentos para acertar'.center(47))
        print(('-'*10).center(47))
        print(' ')

    def trivial_has_acertado(self):
        print(f"¡Enhorabuena, has acertado!".center(47))
        print(f"------- Llevas {self.aciertos} acierto(s) y {self.fallos} fallo(s) -------".center(40))
        print(' ')

    def trivial_has_fallado(self):
        print(f"La respuesta correcta era {self.trivial.get(self.pregunta_random).capitalize()}. Lástima, has fallado esta pregunta".center(40))
        print(f"------- Llevas {self.aciertos} acierto(s) y {self.fallos} fallo(s) -------".center(40))
        print(' ')

    def trivial_pregunta(self):
        print(('-'*10).center(40))
        print('PREGUNTA'.center(40))
        print(self.pregunta_random.center(40))
        print(('-'*10).center(40))

    def ole_ganaste(self):
        print(('-'*48).center(40))
        print('--------HAS GANADO LA PARTIDA ✧⁠\⁠(⁠>⁠o⁠<⁠)⁠ﾉ⁠✧--------'.center(40))
        print(('-'*48).center(40))

    def oh_perdiste(self):
        print(('-'*48).center(40))
        print('---------HAS PERDIDO LA PARTIDA (⁠눈⁠‸⁠눈⁠)----------'.center(40))
        print(('-'*48).center(40))
        
#! ---------------------------------------------------------- Prints: EL AHORCADO
    
    def ahorcado_banner_inicio(self): # inicio del juego + primer display de la palabra oculta
        print(('-'*23).center(40))
        print('EL AHORCADO'.center(40))
        print(('-'*23).center(40))
        print(' ')
        print('Reglas:'.center(40))
        print('Intenta adivinar la palabra secreta'.center(40))
        print('Tienes 7 intentos'.center(40))
        print(' ')
        print('Palabra secreta'.center(40))
        print((self.guiones).center(40))
        print(' ')

    def ahorcado_adivinar_letra(self):
        print(f"-------------- Ronda {self.ronda} --------------".center(40))
        print(f"¡Enhorabuena, has acertado! La letra {self.letra} está en la palabra secreta.".center(40))
        print(f"Las letras utilizadas hasta el momento son: {self.letras_dichas}".center(40))
        print(' ')
        print(f"Palabra secreta:".center(40))
        print(self.guiones.center(40))
        print(' ')

    def ahorcado_fallar_letra(self):
        print(f"-------------- Ronda {self.ronda} --------------".center(40))
        print(f"Lástima, has fallado. La letra {self.letra} no se encuentra en la palabra secreta".center(40))
        print(f"Te quedan {self.vidas} vidas".center(40))
        print(f"Las letras utilizadas hasta el momento son: {self.letras_dichas}".center(40))
        print(' ')
        print(f"Palabra secreta:".center(40))
        print(self.guiones.center(40))
        print(self.ahorcado[self.intentos-1])
        print(' ')
        
#! ---------------------------------------------------------- Prints: PIEDRA, PAPEL Y TIJERA

    def piedrapapeltijera_bienvenida(self):
        print(('-'*37).center(40))
        print(f'PIEDRA, PAPEL O TIJERA'.center(40))
        print(f"Intenta ganar a la máquina".center(40))
        print(f" ")
        print(f"Reglas:".center(40))
        print(f"Papel vence a Piedra".center(40))
        print(f"Piedra vence a Tijera".center(40))
        print(f"Tijeras vence a Papel".center(40))
        print(f"La vencedora será quien consiga 3 victorias".center(40))
        print(f" ") 

    def piedrapapeltijera_empate(self):
        print(f"-------------- Ronda {self.ronda} --------------".center(40))
        print(f"¡Ha habido un empate!".center(40))
        print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
        print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
        print(f" ")

    def piedrapapeltijera_puntomaquina(self):
        print(f"-------------- Ronda {self.ronda} --------------".center(40))
        print(f"¡Lástima! Has perdido...".center(40))
        print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
        print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
        print(f" ")

    def piedrapapeltijera_puntojugadora(self):
        print(f"-------------- Ronda {self.ronda} --------------".center(40))
        print(f"¡Genial! ¡Has ganado!".center(40))
        print(f"Has escogido {self.jugadora} y la máquina {self.maquina}".center(40))
        print(f"---- Jugadora ({self.puntos_jugadora}) vs. Máquina ({self.puntos_maquina}) ---".center(40))
        print(f" ")

    def piedrapapeltijera_ganajugadora(self):
        print(('-'*37).center(40))
        print('---HAS GANADO LA PARTIDA ✧⁠\⁠(⁠>⁠o⁠<⁠)⁠ﾉ⁠✧---'.center(40))
        print(('-'*37).center(40))

    def piedrapapeltijera_ganamaquina(self):
        print(('-'*37).center(40))
        print('----HAS PERDIDO LA PARTIDA (⁠눈⁠‸⁠눈⁠)----'.center(40))
        print(('-'*37).center(40))
```

```python
import random
import string

class Ahorcado(Juegos):
    def __init__ (self):
        self.lista = ['casa', 'perro', 'cachopo', 'dinosaurio', 'python',
              'bucle', 'pesadilla', 'anaconda', 'lista', 'funcion',
              'variable', 'gato', 'lapiz', 'pez', 'sol', 'clavo',
              'pais', 'oro', 'mesa', 'vaso', 'vino', 'cuadro',
              'elefante', 'estrella', 'universo', 'ventana', 'ordenador',
              'quimera']
        self.ahorcado = ['''
                +---+
                |   |
                    |
                    |
                    |
                    |

                =========''', '''
                +---+
                |   |
                O   |
                    |
                    |
                    |
                =========''', '''
                +---+
                |   |
                O   |
                |   |
                    |
                    |
                =========''', '''
                +---+
                |   |
                O   |
               /|   |
                    |
                    |
                =========''', '''
                +---+
                |   |
                O   |
               /|\  |
                    |
                    |
                =========''', '''
                +---+
                |   |
                O   |
               /|\  |
               /    |
                    |
                =========''', '''
                +---+
                |   |
                O   |
               /|\  |
               / \  |
                    |
                =========''']
        self.palabra_secreta = random.choice(self.lista)
        self.guiones = '_ ' * len(self.palabra_secreta)
        self.lista_guiones = self.guiones.split()
        self.vidas = 7
        self.letras_dichas = []
        self.intentos = 0
        self.letra = ''
        
    
    def validar_letra(self): # aquí validamos si la letra es, en efecto, una letra
        self.validacion = False
        while not self.validacion:
            self.letra = input('Introduce una letra').lower()
            if self.letra in string.ascii_lowercase:
                self.letras_dichas.extend(self.letra)
                self.validacion = True
            else:
                return f'{self.letra} no es una opción válida'
            
            
    def adivinar_letra(self):
        
        self.existe = False
        
        if self.letra in self.palabra_secreta:
           for i in range(len(self.palabra_secreta)):
                if self.palabra_secreta[i] == self.letra:
                    self.lista_guiones[i] = self.letra
                    self.guiones = " ".join(self.lista_guiones)
                    self.existe = True     
                    
        if self.existe:
            self.ahorcado_adivinar_letra()
            self.ronda += 1
                
        else:
            self.vidas -= 1
            self.intentos += 1
            self.ahorcado_fallar_letra()
            self.ronda += 1


    def fin_juego(self):
        if self.guiones.replace(' ', '') == self.palabra_secreta:
            print(('-'*48).center(40))
            print('--------HAS GANADO LA PARTIDA ✧⁠\⁠(⁠>⁠o⁠<⁠)⁠ﾉ⁠✧--------'.center(40))
            print(('-'*48).center(40))

        else:
            print(('-'*48).center(40))
            print('---------HAS PERDIDO LA PARTIDA (⁠눈⁠‸⁠눈⁠)----------'.center(40))
            print(('-'*48).center(40))


    def insert_coin(self):
        self.ahorcado_banner_inicio()
        self.ronda = 1
            
        while self.vidas > 0 and self.guiones.replace(' ', '') != self.palabra_secreta:
            self.validar_letra()

            self.adivinar_letra()                      

        self.fin_juego()
        
```

```python
ahorcado = Ahorcado()
```

```python
ahorcado.insert_coin()
```

```python
import random 

class Juego_Piedra_Papel_Tijera(Juegos):
    
    def __init__ (self): 
        
        self.reglas = {"piedra" : "papel" , "papel" : "tijera" , "tijera" : "piedra"} # --En el diccionario de reglas cada clave tiene asociado el valor que la vence en el juego de piedra, papel o tijeras
        self.lista = ['piedra', 'papel', 'tijera'] # --Definimos una lista con las opciones que aceptaremos en la entrada de datos
        self.puntos_jugadora = 0 # --Establecemos los puntos iniciales tanto para la jugadora como para la máquina
        self.puntos_maquina = 0
        self.ronda = 0
        
           
    def entrada_datos (self): # --Método con los inputs necesarios para jugar
        
        self.jugadora = input('Elige opción: Piedra, Papel o Tijera').lower().strip() # --Respuesta de la jugadora (lower para evitar mayúsculas y strip para evitar espacios no deseados antes/después del texto)
        self.maquina = random.choice(self.lista) # --Respuesta de la máquina (random.choice que va a buscar a la lista de elementos posibles definida en el init)
        
    def jugada (self): # --Método con la lógica tras una jugada
        
        if self.jugadora == self.maquina: # --Revisamos si ambas respuestas son iguales para ver si ha habido un empate. Si es así, no se asignarán puntos
            self.piedrapapeltijera_empate()
            
        else:
            
            if self.reglas.get(self.jugadora) == self.maquina: # --Usamos el input de la jugadora como clave para que nos devuelva su valor (recordamos: la opción que la vencería) y si este es igual a la respuesta de la máquina, la máquina gana la ronda
                self.puntos_maquina += 1 # --Como la máquina ha ganado, le sumamos 1 punto
                self.piedrapapeltijera_puntomaquina()
                
            else: # --Si la condición anterior no se cumple, significa que la jugadora ha ganado, ya que la máquina no ha obtenido la opción que podría vencerla
                self.puntos_jugadora += 1 # --Como la jugadora ha ganado, le sumamos 1 punto
                self.piedrapapeltijera_puntojugadora()
                
    def resultado (self): # --Método con la definición del resultado del juego
        
        if self.puntos_jugadora > self.puntos_maquina:
            self.piedrapapeltijera_ganajugadora()
            
        else:
            self.piedrapapeltijera_ganamaquina()
            
        
    def jugar (self):
        
        self.piedrapapeltijera_bienvenida()
        
        while self.puntos_jugadora < 3 and self.puntos_maquina < 3:
            
            self.ronda += 1
        
            self.entrada_datos()
            
            if self.jugadora not in self.reglas:
                print('¡Vaya! Ha ocurrido un error. Prueba otra vez con Piedra, Papel o Tijera'.center(40))
                print(f" ")
                
            else: 
                self.jugada()
                
        self.resultado() 
```

```python
pipati = Juego_Piedra_Papel_Tijera()
```

```python
pipati.jugar()
```

```python
import random

class Trivial(Juegos):
    def __init__(self):

        self.trivial = {'¿Cuál es la capital de Qatar?' : 'doha' , 
           '¿Cuál es el país más grande del mundo?' : 'rusia', 
           '¿Cuál es la montaña más alta del mundo?' : 'everest', 
           '¿Cuántos estados hay en Estados Unidos?' : '50', 
           '¿Cuántos países forman parte de la Unión Europea?' : '27', 
           '¿Cuál es el continente más poblado?' : 'asia', 
           '¿Cuál es el río más largo de Europa?' : 'volga', 
           '¿Cuál es el país menos densamente poblado del mundo?' : 'mongolia', 
           '¿En qué país latinoamericano encontramos la ciudad de Cochabamba?' : 'bolivia', 
           '¿Qué país africano tiene como capital Nairobi?' : 'kenia'}
        self.lista_preguntas = list(self.trivial.keys())
        self.aciertos = 0
        self.intentos = 2
        self.fallos = 0       
    

       
    def jugada(self):
        
        if self.intentos > 0:
            
            if self.intentos == 2: 
                self.ronda +=1
                print(f"------------------- Ronda {self.ronda} --------------------".center(40))
                self.pregunta_random = random.choice(self.lista_preguntas)
                self.lista_preguntas.remove(self.pregunta_random) # ----- eliminamos el random de la lista de preguntas
                # --Mantiene la misma pregunta durante los 3 intentos para acertarla
            
            respuesta = input(self.pregunta_random).lower().strip() # ----- capitalize.() para que salgan correctamente escritas en el print de lín. 45
            # --Solicita la respuesta de la jugadora

            if self.trivial.get(self.pregunta_random) == respuesta:
                self.aciertos += 1
                self.intentos = 2
                self.trivial_has_acertado()
                # --Si la respuesta es correcta, da un punto y pasa a la siguiente pregunta. También se restablece el número de intentos para la siguiente ronda
            
            else: 
                self.intentos -= 1
                if self.intentos != 0: # --Condicional extra para que el print no muestre un mensaje diciendo que quedan "0 intentos"
                    print(f"No es correcto, pero aún te quedan {self.intentos} intento(s). ¡Prueba de nuevo!")
                    # --Si la respuesta es errónea, pero aún quedan intentos, vuelve a solicitar respuesta a la jugadora. Cada respuesta errónea resta 1 intento  
        else:
            self.fallos += 1
            self.intentos = 2
            self.trivial_has_fallado()
    
    def fin_juego(self):
        if self.aciertos > 4:
            self.ole_ganaste()

        elif self.fallos > 2:
            self.oh_perdiste()

    def insert_coin(self):
        self.trivial_banner_inicio()
        
        self.ronda = 0
        
        while self.aciertos < 5 and self.fallos < 3:
            
            self.jugada()
        
        self.fin_juego()
```
